```python
pollutant=input('please write either NO2, O3, or PM2.5')

if pollutant=='NO2':
  airquality2=airquality.loc[airquality['pollutant']=='NO2',['country', 'value']]
  grouped1=airquality2.groupby(["country"], as_index=False)["value"].mean()
  fig=plt.figure()
  ax=fig.add_subplot()
  no2=ax.plot(grouped1['country'], grouped1['value'])
  ax.set_title("Average Level of Pollutant by Country (2023-2024)")
  ax.set_xlabel('Country')
  ax.set_ylabel('NO2 Concentration (micrograms per cubic meter)')
  plt.xticks(rotation=90)
elif pollutant=='O3':
  airquality3=airquality.loc[airquality['pollutant']=='O3', ['country', 'value']]
  grouped2=airquality3.groupby(['country'], as_index=False)['value'].mean()
  fig=plt.figure()
  ax=fig.add_subplot()
  o3=ax.plot(grouped2['country'], grouped2['value'])
  ax.set_title("Average Level of Pollutant by Country (2023-2024)")
  ax.set_xlabel('Country')
  ax.set_ylabel('O3 Concentration (micrograms per cubic meter)')
  plt.xticks(rotation=90)
else:
  airquality4=airquality.loc[airquality['pollutant']=='PM2.5', ['country', 'value']]
  grouped3=airquality4.groupby(['country'], as_index=False)['value'].mean()
  fig=plt.figure()
  ax=fig.add_subplot()
  pm2_5=ax.plot(grouped3['country'], grouped3['value'])
  ax.set_title("Average Level of Pollutant by Country (2023-2024)")
  ax.set_xlabel('Country')
  ax.set_ylabel('PM2.5 Concentration (micrograms per cubic meter)')
  plt.xticks(rotation=90)
```
